```python
def fit_and_predict(modelName, model, train_x, train_y): 
    k = 10
    scores = cross_val_score(model, train_x, train_y, cv = k)
    hit_rate = np.mean(scores)
    msg = "Hit Rate from {}: {}".format(modelName, hit_rate)
    print(msg)
    return hit_rate
```

```python
X_train_test = X_train_robust.values
X_to_predict = X_train_to_predict.values
Y_labels = Y_labels.values
```

```python
from sklearn.model_selection import train_test_split
train_data_X ,test_data_X,train_data_Y,test_data_Y = train_test_split(X_train_test,Y_labels,test_size = 0.2, random_state=42)
```

```python
results = {}
win_odds = [0]*len(df_to_predict)
which={}
for i in range(0, len(df_to_predict)):
    which[i] = []
```

```python
def predict_results(model, result):
    results[model] = result
    model.fit(train_data_X, train_data_Y)
    resultOscar = model.predict(X_to_predict)
    probs = model.predict_proba(X_to_predict)
    win_probs = []
    for i in probs:
        win_probs.append(i[1])
    print('Result: ' + str(resultOscar))
    flag = 0
    for i in resultOscar:
        if i == 1:
            flag = 1
    if flag == 0:
        max_index = win_probs.index(max(win_probs))
        result = [1 if i == max_index else 0 for i in range(len(probs))]
        for i in range(0,len(resultOscar)):
                if(result[i] == 1.0):
                    win_odds[i] = win_odds[i] + 1
                    which[i].append(model)
    else:
        for i in range(0,len(resultOscar)):
                if(resultOscar[i] == 1.0):
                    win_odds[i] = win_odds[i] + 1
                    which[i].append(model)
```
